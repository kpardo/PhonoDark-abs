```python
import matplotlib.pyplot as plt


import pda.material
import pda.rate as r
import pda.reach as re
import pda.couplings as coup
from pda.plotting import *
from pda.constants import *

%matplotlib inline
```

```python
## First, set the DM masses you want to probe
mlist = np.logspace(-2,np.log10(0.5),int(1e5));
```

```python
## Now set your material.
## GaAs is a simple one that runs quickly.
mat = material.Material('GaAs')
```

```python
## Now set the coupling you want for your DM.
c = coup.ScalarE(omega=mlist, mat=mat)
```

```python
## Now calculate the rate.
## Make sure to multiply in the target mass and time 
## -- in this case 1 KG-YR
r.rate(mass_list=mlist, mat=mat, coupling=c)*KG_YR
```

```python
## Finally, you can plot the reach.
fig, axes = plt.subplots(1,1)
colors = ['firebrick']
plot_coupling(axes, [c], colors)
## and finish this off by making it prettier
set_custom_tick_options(axes, major_right = False, minor_right = False)
set_custom_axes(axes, 'x', 1, 3,
           ax_type = 'log', 
           label = r'$m_\phi \, [\mathrm{meV}]$',
           show_first = True)
set_custom_axes(axes, 'y', 7, 11,
           ax_type = 'log', 
           label = r'$d_e$', 
           show_first = False)
```
